```python
# Arquivos e suas respectivas origens
arquivos = {
    'AliExpress': 'data/raw_data/Meganium_Sales_Data_-_AliExpress.csv',
    'Etsy':       'data/raw_data/Meganium_Sales_Data_-_Etsy.csv',
    'Shopee':     'data/raw_data/Meganium_Sales_Data_-_Shopee.csv'
}

# Lista para armazenar os DataFrames
dataframes = []

# Carrega cada arquivo, adicionando uma coluna 'origem'
for origem, caminho in arquivos.items():
    df = pd.read_csv(caminho)
    df['origem'] = origem
    dataframes.append(df)

# Concatena todos em um Ãºnico DataFrame
df_total = pd.concat(dataframes, ignore_index=True)

## formatar datas do formato americano (YYYY-MM-DD) para o brasileiro (DD/MM/YYYY)

df_total['date'] = pd.to_datetime(df_total['date'], errors='coerce')
df_total['buyer_birth_date'] = pd.to_datetime(df_total['buyer_birth_date'], errors='coerce')

# Converte para o formato brasileiro
df_total['date'] = df_total['date'].dt.strftime('%d/%m/%Y')
df_total['buyer_birth_date'] = df_total['buyer_birth_date'].dt.strftime('%d/%m/%Y')

df = df_total.copy()


# Visualiza os dados combinados
df_total.head(30)
```

```python
## Produto mais vendido

produto_mais_vendido = df.groupby("product_sold")["quantity"].sum().sort_values(ascending=False)
print("Produto mais vendido:")
print(produto_mais_vendido.head(1))

# GrÃ¡fico dos top 5 produtos mais vendidos
produto_mais_vendido.head(5).plot(kind="bar", title="Top 5 produtos mais vendidos", ylabel="quantity")
plt.show()
```

```python
## Site que mais vendeu

site_que_mais_vendeu = df.groupby("site")["quantity"].sum().sort_values(ascending=False)
print(site_que_mais_vendeu.head(1).to_frame())
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Consolidar os dados
arquivos = {
    'AliExpress': 'data/raw_data/Meganium_Sales_Data_-_AliExpress.csv',
    'Etsy': 'data/raw_data/Meganium_Sales_Data_-_Etsy.csv',
    'Shopee': 'data/raw_data/Meganium_Sales_Data_-_Shopee.csv'
}

dfs = []
for origem, caminho in arquivos.items():
    df = pd.read_csv(caminho)
    df['origem'] = origem
    dfs.append(df)

df = pd.concat(dfs, ignore_index=True)

# Garantir datas em datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['buyer_birth_date'] = pd.to_datetime(df['buyer_birth_date'], errors='coerce')

# Criar coluna de receita total
df['receita'] = df['price'] * df['quantity']

# Produtos mais vendidos
top_produtos = df.groupby("product_sold")["quantity"].sum().sort_values(ascending=False)
print("Top 5 produtos vendidos:")
print(top_produtos.head(5))

# GrÃ¡fico
top_produtos.head(5).plot(kind="bar", title="Top 5 Produtos Mais Vendidos")
plt.ylabel("Quantidade")
plt.show()

# Receita por paÃ­s
receita_pais = df.groupby("ship_to_country")["receita"].sum().sort_values(ascending=False)
print("Receita total por paÃ­s:")
print(receita_pais)

# Produtos mais populares por paÃ­s
popular_por_pais = df.groupby(['ship_to_country', 'product_sold'])['quantity'].sum()
popular_por_pais = popular_por_pais.reset_index()
mais_popular = popular_por_pais.sort_values(['ship_to_country', 'quantity'], ascending=[True, False]).drop_duplicates('ship_to_country')
print("Produto mais popular por paÃ­s:")
print(mais_popular[['ship_to_country', 'product_sold', 'quantity']])

# Receita mÃ©dia por marketplace
receita_origem = df.groupby("origem")["receita"].mean()
print("Receita mÃ©dia por origem:")
print(receita_origem)

# SugestÃ£o para logÃ­stica (volume por paÃ­s)
volume_por_pais = df['ship_to_country'].value_counts()
print("Volume de pedidos por paÃ­s:")
print(volume_por_pais)
```
